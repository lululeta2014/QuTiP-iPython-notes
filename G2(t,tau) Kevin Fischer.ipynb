# Two time correlations in QuTiP for time dependent Hamiltonians

#### Author: Kevin Fischer, group of Jelena Vuckovic, Stanford University

Application of the quantum regression theorem to the problem of calculating the two time correlation

$$ \left< A(t)B(t+\tau)C(t+\tau)D(t) \right> $$

yields the following result:

$$ \left< A(t)B(t+\tau)C(t+\tau)D(t) \right> = \text{Tr}_\text{sys}\{B\,C\,\Lambda(t,t+\tau)\} $$

where $\Lambda(t,t+\tau)$ is governed by the following evolution equation

$$ \partial_\tau \Lambda(t,t+\tau) = \mathcal{L}(t+\tau) \Lambda(t,t+\tau) $$

subject to the initial condition

$$ \Lambda(t,t) = D \rho(t) A $$

```python
from qutip import *
from pylab import *
from scipy import *
from qutip.odechecks import _ode_checks
```

## Example implementation

I have attempted to implement this correlator in keeping with QuTiP conventions. It accepts time dependent Hamiltonians in either cython or fn list format. Additionally, it behaves like <i>correlation_4op_2t</i> when a time independent Hamiltonian is entered.

```python
from re import sub

def _transform_H_me_4op_2t_td(H, args={}):
    """
    Time shift the hamiltonian, as required for the quantum regression theorem
    """
    
    args_new = args.copy()
    args_new["t0_correlation"] = 0

    if isinstance(H, Qobj):
        # constant hamiltonian
        return H, args
    
    if isinstance(H, types.FunctionType):
        # old style time-dependence
        raise TypeError("This style of specifying time-dependence is deprecated")    
    
    if isinstance(H, list):
        # determine if we are dealing with list of [Qobj, string] or
        # [Qobj, function] style time-dependences (for pure python and
        # cython, respectively)
        H_new = []
        
        for i in range(len(H)):
            if isinstance(H[i],list):
                # modify hamiltonian time depence in accordance with the quantum
                # regression theorem
                if isinstance(H[i][1], types.FunctionType):
                    fn = lambda t,args: H[i][1](t+args["t0_correlation"],args)
                else:
                    fn = sub("(?<=[^0-9a-zA-Z_])t(?=[^0-9a-zA-Z_])",\
                             "t + t0_correlation",H[i][1])
                
                H_new.append([H[i][0],fn])
            else:
                H_new.append(H[i])
        
        return H_new, args_new

def _correlation_me_4op_2t_td(H, rho0, tlist, taulist, c_ops,
                           a_op, b_op, c_op, d_op, reverse=False, args={},
                           options=Odeoptions()):
    """
    Calculate the four-operator two-time correlation function on the form
    <A(t)B(t+tau)C(t+tau)D(t)>.

    See, Gardiner, Quantum Noise, Section 5.2.1
    """

    if debug:
        print(inspect.stack()[0][3])

    if rho0 is None:
        raise TypeError("Missing initial state")
    elif rho0 and isket(rho0):
        rho0 = ket2dm(rho0)

    C_mat = np.zeros([np.size(tlist), np.size(taulist)], dtype=complex)

    rho_t = mesolve(H, rho0, tlist, c_ops, [], args=args, options=options).states
    
    H_new, args_new = _transform_H_me_4op_2t_td(H, args)
    
    for t_idx, rho in enumerate(rho_t):
        args_new["t0_correlation"] = tlist[t_idx]
        C_mat[t_idx, :] = mesolve(H_new, d_op * rho * a_op, taulist,
                                  c_ops, [b_op * c_op], args=args_new,
                                  options=options).expect[0]
        if t_idx == 1:
            options.rhs_reuse = True

    return C_mat

def correlation_4op_2t_td(H, rho0, tlist, taulist, c_ops, a_op, b_op, c_op, d_op,
                       solver="me", args={}, options=Odeoptions()):
    """
    Calculate the four-operator two-time correlation function on the from
    :math:`\left<A(t)B(t+\\tau)C(t+\\tau)D(t)\\right>` using the quantum
    regression theorem and the solver indicated by the 'solver' parameter.

    Parameters
    ----------

    H : :class:`qutip.qobj.Qobj`
        system Hamiltonian, or a callback function for time-dependent Hamiltonians.

    rho0 : :class:`qutip.qobj.Qobj`
        Initial state density matrix (or state vector). If 'rho0' is
        'None', then the steady state will be used as initial state.

    tlist : *list* / *array*
        list of times for :math:`t`.

    taulist : *list* / *array*
        list of times for :math:`\\tau`.

    c_ops : list of :class:`qutip.qobj.Qobj`
        list of collapse operators. (does not accept time dependence)

    a_op : :class:`qutip.qobj.Qobj`
        operator A.

    b_op : :class:`qutip.qobj.Qobj`
        operator B.

    c_op : :class:`qutip.qobj.Qobj`
        operator C.

    d_op : :class:`qutip.qobj.Qobj`
        operator D.

    solver : str
        choice of solver (currently only `me` for master-equation)

    Returns
    -------

    corr_mat: *array*
        An 2-dimensional *array* (matrix) of correlation values for the times
        specified by `tlist` (first index) and `taulist` (second index). If
        `tlist` is `None`, then a 1-dimensional *array* of correlation values
        is returned instead.

    References
    ----------

    See, Gardiner, Quantum Noise, Section 5.2.1.

    """
    
    if rho0 is None and not isinstance(H, list):
        rho0 = steadystate(H, c_ops)
    elif rho0 and isket(rho0):
        rho0 = ket2dm(rho0)

    if debug:
        print(inspect.stack()[0][3])

    if solver == "me":
        return _correlation_me_4op_2t_td(H, rho0, tlist, taulist, c_ops,
                                      a_op, b_op, c_op, d_op, args=args,
                                      options=options)
    else:
        raise NotImplementedError("Unrecognized choice of solver %s." % solver)
```

## Consider two time correlations of a driven two level system (TLS)

Suppose we wish to calculate time dependent photon intensity correlations for a driven two level system with arbitrary pulse length, i.e.

$$ G^2(t,\tau)= \gamma^2 \langle \hat{\sigma}(t)^\dagger \hat{\sigma}(t+\tau)^\dagger \hat{\sigma}(t+\tau) \hat{\sigma}(t) \rangle$$

We identify this correlation as of the form 

$$ \left< A(t)B(t+\tau)C(t+\tau)D(t) \right> $$

```python
# Generate TLS with following system parameters
gamma = 1
tp = 2 # guassian pulse parameter
E0 = 1.5
t_offset = 2*tp
tmax = 2*t_offset + 3/gamma
npts = 200

sm = destroy(2)
c_ops = [sqrt(gamma)*sm]

psi0 = fock(2,0)
rho0 = ket2dm(psi0)

# <A(t)B(t+tau)C(t+tau)D(t)>
a_op = sm.dag()
b_op = sm.dag()
c_op = sm
d_op = sm
```

## Two time correlation test with time independent Hamiltonian

Nothing special here, just a test to show it works as the old <i>correlation_4op_2t</i> would for a time independent Hamiltonian.

```python
H_ss = E0*(sm + sm.dag())
```

```python
tlist_ss = linspace(0,5*gamma,npts)
taulist_ss = linspace(0,5*gamma,npts)

corr_ss = correlation_4op_2t_td(H_ss, rho0, tlist_ss, taulist_ss, c_ops, a_op, b_op, c_op, d_op,
                       solver="me", options=Odeoptions())
```

```python
figure(figsize=(7,5))
pcolor(taulist_ss/gamma,tlist_ss/gamma,abs(corr_ss))
xlim([min(taulist_ss/gamma),max(taulist_ss/gamma)]); ylim([min(tlist_ss/gamma),max(tlist_ss/gamma)])
xlabel(u'$\\tau/\gamma$',fontsize='18'); ylabel(u'$t/\gamma$',fontsize='18')
title(u'$G^2(t,\\tau)$ for time independent $H$',fontsize='18')
```

## Two time correlation test with cython td format

Now we test for the case of a driving pulse, for both cython and fn list td format

```python
args = {"t_off":t_offset,"tp":tp}

H_cy = [[E0*(sm + sm.dag()),"exp(-(t-t_off)**2/(2*tp**2))"]]
```

```python
tlist = linspace(0,tmax,npts)
taulist = linspace(0,tmax,npts)

corr_cy = correlation_4op_2t_td(H_cy, rho0, tlist, taulist, c_ops, a_op, b_op, c_op, d_op,
                       solver="me", args=args, options=Odeoptions())
```

```python
figure(figsize=(7,5))
pcolor(taulist/gamma,tlist/gamma,abs(corr_cy))
xlim([min(taulist/gamma),max(taulist/gamma)]); ylim([min(tlist/gamma),max(tlist/gamma)])
xlabel(u'$\\tau/\gamma$',fontsize='18'); ylabel(u'$t/\gamma$',fontsize='18')
title(u'$G^2(t,\\tau)$ for $H(t)$ with cython td format',fontsize='18')
```

## Two time correlation test with function td format

```python
args = {"t_off":t_offset,"tp":tp}

def pulse_shape(t, args):
    return exp(-(t-args["t_off"])**2/(2*args["tp"]**2))

H_fn = [[E0*(sm + sm.dag()),pulse_shape]]
```

```python
tlist = linspace(0,tmax,npts)
taulist = linspace(0,tmax,npts)

corr_fn = correlation_4op_2t_td(H_fn, rho0, tlist, taulist, c_ops, a_op, b_op, c_op, d_op,
                       solver="me", args=args, options=Odeoptions())
```

```python
figure(figsize=(7,5))
pcolor(taulist/gamma,tlist/gamma,abs(corr_fn))
xlim([min(taulist/gamma),max(taulist/gamma)]); ylim([min(tlist/gamma),max(tlist/gamma)])
xlabel(u'$\\tau/\gamma$',fontsize='18'); ylabel(u'$t/\gamma$',fontsize='18')
title(u'$G^2(t,\\tau)$ for $H(t)$ with function td format',fontsize='20')
```

## Single time correlation $G^2(\tau)$ for $H(t)$

Integrating over $t$, we arrive at the single time correlation function that describes net correlations between pairs of photons in the wavepacket

$$G^2(\tau)= \gamma^2 \int dt \, \langle \hat{\sigma}(t)^\dagger \hat{\sigma}(t+\tau)^\dagger \hat{\sigma}(t+\tau) \hat{\sigma}(t) \rangle$$

```python
G2_t_tau = corr_fn*gamma
dt = float(tmax)/(npts-1)
G2_tau = dt*sum(G2_t_tau,0)
```

```python
figure(figsize=(7,5))
plot(taulist/gamma,abs(G2_tau))
xlim([min(taulist/gamma),max(taulist/gamma)])
xlabel(u'$\\tau/\gamma$',fontsize='18'); title(u'$G^2(\\tau)$ for $H(t)$',fontsize='20')
```

## Experimental zero delay autocorrelation $g^2(0)$ for $H(t)$

To model the detection process, the zero delay correlation is given by

$$ G^2(0)= \langle m(m-1) \rangle $$

where the expectation value is over a measured photocount distribution $P_m(T)$. We can calculate this distribution and thus the zero delay correlation directly using the QuTiP Monte Carlo solver to monitor collapse events. Furthermore, we define the normalized version as

$$ g^2(0)= \frac{\langle m(m-1) \rangle} {\langle m \rangle^2} $$

```python
from IPython.display import display, Math, Latex
import os, sys

tlist_mc = linspace(0,tmax,2)

devnull = open(os.devnull, 'w') # suppress the output from mcsolve which doesn't interface nicely with iPython
sys.stdout, sys.stderr = devnull, devnull # in the version
result_mc = mcsolve(H_cy, psi0, tlist_mc, c_ops, [sm.dag()*sm], ntraj=10000, args=args);
sys.stdout = sys.__stdout__ # return system output pointers
sys.stderr = sys.__stderr__ # return system output pointers

# bin the collapse events to generate Pm
ncollapse = [result_mc.col_times[i].size for i in range(result_mc.col_times.size)] 
bns = range(60)
Pm = numpy.histogram(ncollapse, bins=bns, density=True)[0]
    
g20_mc = sum([p*m*(m-1) for m,p in enumerate(Pm)])/sum([p*m for m,p in enumerate(Pm)])**2

display(Math(r'g^2_\text{MC}(0) = ' + str(round(g20_mc,2))))
```

We can make the connection to the instantaneous output field flux correlations (from the TLS emission) through the corresponding quantum mechanical photocount operator, given by

$$ \hat{M}_\text{D} =\gamma \int_\text{D} dt \, \hat{\sigma}(t)^\dagger \hat{\sigma}(t) $$

such that

$$ \langle m \rangle = \langle \hat{M}_\text{D} \rangle $$

Applying this time integrated operator to the definition of $G^2(0)$, we arrive at

$$G^2(0)= \gamma^2 \iint_\text{D} dt \, d(t+\tau) \: \langle \hat{\sigma}(t)^\dagger \hat{\sigma}(t+\tau)^\dagger \hat{\sigma}(t+\tau) \hat{\sigma}(t) \rangle$$

where the normalized version is

$$g^2(0) = \frac{G^2(0)} {\left(\gamma \langle m \rangle \right)^2} = \frac{G^2(0)} {\left( \gamma \int_\text{D} dt \, \langle \hat{\sigma}(t)^\dagger \hat{\sigma}(t) \rangle\right)^2}$$

```python
exp_M_D = sum(mesolve(H_fn, rho0, tlist, c_ops, [sm.dag()*sm], args=args).expect[0])
exp_n = sqrt(gamma)*dt*exp_M_D
g20 = abs(2*sum(G2_tau)*dt/exp_n**2) # factor of 2 from negative time correlations

display(Math(r'g^2_\text{ME}(0) = ' + str(round(g20,2))))
```

Which agrees very well with the Monte Carlo value.
